```python
#import library yang dibutuhkan 

import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

## Import Data

```python
!wget http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx
```

```python
df = pd.read_excel('Online Retail.xlsx')
df.head(20)
```

```python
df.describe()
```

## Cleaning Data

```python
#data pada kolom Description yang mengandung strip diawal dan diakhir dihapus
df['Description'] = df['Description'].str.strip()
# data pada kolom InvoiceNo yang mengandung nilai NA maka akan dihapus
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True) 
# data pada kolom InvoiceNo diubah menjadi string
df['InvoiceNo'] = df['InvoiceNo'].astype('str') 
# data pada kolom InvoiceNo yang mengandung karakter C maka akan dihapus
df = df[~df['InvoiceNo'].str.contains('C')]
# tampilkan data
df.head()
```

```python
datanegara = df.groupby(['Country']).count()
datanegara.sort_values(by='Quantity', ascending=False)
```

## Filtering Data

```python
basket = (df[df['Country'] =="France"]           
          .groupby(['InvoiceNo', 'Description'])['Quantity']           
          .sum().unstack().reset_index().fillna(0)           
          .set_index('InvoiceNo'))
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(encode_units)
 
#menghapus kolom postage
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head()
```

```python
# Build up the frequent items 
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
frequent_itemsets.head()
```

```python
# Create the rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
filtered_rules = rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]

filtered_rules.sort_values(by=['support','confidence'], ascending=False)
```

Data filtered_rules telah dilakukan sorting dari nilai support dan confidence yang tertinggi ke terendah.

Support merupakan nilai penunjang yaitu persentase kombinasi item dalam database dan confidence merupakan nilai kepastian yaitu kuatnya hubungan antar item dalam aturan asosiatif. 

{(SET/6 RED SPOTTY PAPER PLATES), 	(SET/6 RED SPOTTY PAPER CUPS)} -> {support = 0.12 atau 12 % , confidence = 0.96 atau 96 %}. 

Pernyataan diatas merupakan nilai support dan nilai confidence tertinggi pada item SET/6 RED SPOTTY PAPER PLATES dan SET/6 RED SPOTTY PAPER CUPS di database. 

Artinya konsumen membeli item SET/6 RED SPOTTY PAPER PLATES  dengan kemungkinan 12% dan memiliki kemungkinan membeli SET/6 RED SPOTTY PAPER PLATES lalu juga membeli  SET/6 RED SPOTTY PAPER CUPS dengan kemungkinan 96 %.
